# Removing Outliers

## Using IQR SCORE

```python
df=weather[[' _hail',' _rain',' _snow',' _thunder', ' _tornado', ' _fog' ]]
```

```python
weather=weather.drop([' _wdire',' _conds',' _hail',' _rain',' _snow',' _thunder',' _tornado',' _fog'],axis=1)
```

```python
Q1=weather.quantile(0.05)
```

```python
Q3=weather.quantile(0.95)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
weather_refined = weather[~((weather < (Q1 - 1.5 * IQR)) |(weather > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
weather_refined.head()
```

```python
weather_refined.shape
```

```python
weather_refined.plot(figsize=(12,10))
```

# Make pie charts for different weather conditions in a month

```python
weather_df=pd.read_csv("C:\\Users\\Aashita Arora\\Documents\\work\\weather.csv")
```

```python
date = pd.to_datetime(weather_df['datetime_utc'])
weather_df.insert(0,'Year',date.dt.year)
weather_df.insert(1,'Month',date.dt.month)
weather_df.insert(2,'Day',date.dt.day)
weather_df.insert(3,'Hour',date.dt.hour)
```

```python
weather_df=weather_df.drop(['datetime_utc'],axis=1)
```

```python
wsub=weather_df[['Month',' _conds']]
```

```python
w=wsub.groupby('Month')
```

```python
month_dict={1:'Jan',2:'Feb',3:'March',4:'April',5:'May',6:'June',7:'July',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
```

```python
colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6']
```

```python
plt.subplots(figsize=(50,50))
count = int(1)

for i,group in w:
    plt.subplot(4,3,count)  
    plt.title(month_dict[i])
    month_data = group[' _conds'].value_counts(sort=True, ascending=False)
    explode = (0.1,0,0,0,0,0)
    
    #Selecting top 6 values as otherwise many low values will also be plotted
    labels = month_data.index.tolist()[:5]
    sizes = month_data.tolist()[:5]
    
    #Adding all low values as title others in the pie-chart
    labels.append('Others')
    sizes.append(sum(month_data.tolist()[6:]))
    
    plt.pie(sizes, labels=labels,colors=colors, autopct='%1.1f%%',shadow=True, explode=explode)
    count = count+1
plt.show()
```
